# 2. Conclusion

## 2.1. Summary of Findings + Applications for Bike Sharing Companies  
We gather from the workings of both models that various categories exert different amounts of influence on the total count variable. However, these levels vary depending on the combinations they occur in which then affects the total count differently on different days. 

We have seen that we make use of highly correlated categories to predict `total_count`, which provides useful insights for bike sharing companies. 

For example, we saw previously that the variable `is_peak_hour` had one of the highest correlations with total count from 7 to 9am, and 4 to 6pm on the weekdays. These hours had the highest spike in bike users, and this data-driven insight can help bike sharing companies to allocate a greater number of bikes during this time frame.

Our chosen **MLP Model 2.2** accounts for the ever changing weights of variables and would allow bike sharing companies to allocate enough bikes to accommodate any potential crowd for that day. This would provide these companies the opportunity to explore different pricing strategies based on demand to maximise their profits. 

Ultimately, we are able to help bike sharing companies predict accurately the number of bikes that would be rented at any hour of the day, based on several features. :-)


## 2.2. Limitations
With that said, one limitation of our models are that they did not account for all variables and neglected those with weak correlations, choosing to work with only significant categorical variables decided from dummy coding. Hence, one future consideration is to use all variables in the modelling, as both models chosen can identify the significant variables themselves.

Additionally, the reverted values still have significant errors, which is something we could work on in the future to reduce.
